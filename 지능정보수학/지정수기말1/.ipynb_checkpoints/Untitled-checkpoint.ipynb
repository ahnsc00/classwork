```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def euclideanDist(x, sample):
    return np.sqrt(np.power(x-sample, 2).sum(axis=1)).sort_values(ascending=True)
```

```python
def manhattanDist(x, sample):
    return np.absolute(x-sample).sum(axis=1).sort_values(ascending=True)
```

```python
def knn(X, Y, sample, k, distance):
    pred = []
    x = X.copy()
    y = Y.copy()
    for i in sample.index:
        if (distance == 'Euclidean'):
            dist = euclideanDist(x, sample.iloc[i])
        elif (distance == 'Manhattan'):
            dist = manhattanDist(x, sample.iloc[i])
        else:
            return 0
        classification = y.iloc[dist[0:k].index].value_counts().reset_index().iloc[0,0]
        pred.append(classification)
        x.loc[len(x)] = sample.iloc[i]
        y.loc[len(y)] = [classification]
    return pred
```

```python
draft = pd.read_csv('./Speed-Agility-Draft.csv')

X = draft[['Speed', 'Agility']]
Y = draft[['Draft']]

sample = pd.DataFrame([[6.75, 3], [5.34, 6], [4.67, 8.4], [7, 7], [7.8, 5.4]], columns = ['Speed', 'Agility'])
```

```python
k = 3
sample['3_Euclidean'] = knn(X, Y, sample, k, 'Euclidean')
sample['3_Manhattan'] = knn(X, Y, sample, k, 'Manhattan')

k = 5
sample['5_Euclidean'] = knn(X, Y, sample, k, 'Euclidean')
sample['5_Manhattan'] = knn(X, Y, sample, k, 'Manhattan')
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d

vor = Voronoi(pd.concat([X, sample[['Speed', 'Agility']]]))
```

```python
ax = sns.scatterplot(x = 'Speed', y = 'Agility',hue = 'Draft', s=80, data = draft)
plt.show()
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d

vor = Voronoi(pd.concat([X, sample[['Speed', 'Agility']]]))

ax = voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=2, line_alpha=0.6, point_size=0)
ax = sns.scatterplot(x = 'Speed', y = 'Agility',hue = 'Draft', s=50, data = draft)
ax = sns.scatterplot(x = 'Speed', y = 'Agility', hue = '3_Euclidean',marker = "X", s=200, data = sample, legend = False)
plt.show()
```

```python
ax = voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=2, line_alpha=0.6, point_size=0)
ax = sns.scatterplot(x = 'Speed', y = 'Agility',hue = 'Draft', s=50, data = draft)
ax = sns.scatterplot(x = 'Speed', y = 'Agility', hue = '3_Manhattan',marker = "X", s=200, data = sample, legend = False)
plt.show()
```

```python
ax = voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=2, line_alpha=0.6, point_size=0)
ax = sns.scatterplot(x = 'Speed', y = 'Agility',hue = 'Draft', s=50, data = draft)
ax = sns.scatterplot(x = 'Speed', y = 'Agility', hue = '5_Euclidean',marker = "X", s=200, data = sample, legend = False)
plt.show()
```

```python
ax = voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=2, line_alpha=0.6, point_size=0)
ax = sns.scatterplot(x = 'Speed', y = 'Agility',hue = 'Draft', s=50, data = draft)
ax = sns.scatterplot(x = 'Speed', y = 'Agility', hue = '5_Manhattan',marker = "X", s=200, data = sample, legend = False)
plt.show()
```

```python
result = pd.concat([draft, sample[['Speed', 'Agility', '5_Euclidean']].rename(columns = {'5_Euclidean':'Draft'})])
result = result[['Speed', 'Agility', 'Draft']]
np.round(result[result['Draft'] == True].mean(), 2)
```
